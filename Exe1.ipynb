## Tarea 1 -> Regresiones Lineales, Interpolaci√≥n Lineal

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([3, 6, 9, 12, 15, 18])
y1 = np.array([14, 20, 26, 32, 38, 44])
plt.plot(x1, y1,'-b')
```

```python
x2 = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y2 = np.array([21, 35, 57, 87, 125, 171, 225, 287])
plt.plot(x2, y2,'-r')
```

```python
x3 = np.array([1.5, 2.5, 3.5, 4.5, 5.5, 6.5, 7.5, 8.5])
y3 = np.array([-0.5, 4.5, 9.5, 14.5, 19.5, 24.5, 29.5, 34.5])
plt.plot(x3, y3,'-g')
```

```python
#Primer Dataframe - Interpolaci√≥n lineal
x11=3
y11=14
x21=9
y21=26
xa1=7
ya1 = ((y21-y11)/(x21-x11))*(xa1-x11)+y11

x12=6
y12=20
x22=12
y22=32
xa2=11
ya2 = ((y22-y12)/(x22-x12))*(xa2-x12)+y12

x13=15
y13=38
x23=18
y23=44
xa3=17
ya3 = ((y23-y13)/(x23-x13))*(xa3-x13)+y13

x14=3
y14=14
x24=18
y24=44
xa4=16
ya4 = ((y24-y14)/(x24-x14))*(xa4-x14)+y14

x15=9
y15=26
x25=12
y25=32
xa5=10
ya5 = ((y25-y15)/(x25-x15))*(xa5-x15)+y15

x1 = np.append(x1, [xa1, xa2, xa3, xa4, xa5])
y1 = np.append(y1, [ya1, ya2, ya3, ya4, ya5])
plt.plot(x1, y1,'-b')
```

```python
#Modelos lineales
x11=3
y11=14
x21=9
y21=26

m = (y21-y11)/(x21-x11)
# y = mx + b -> b = y - (mx)
b = 26-(m*14)
# yn = mx + b
yn = m*x1 + b

plt.plot(x1, y1, '-g')
```

```python
#Metricas de error
Error=np.sqrt(sum((y1-yn)**2)/len(y1))
print(Error)

plt.plot(x1, y1, '-g')
plt.plot(x1, yn, '-b')
```
